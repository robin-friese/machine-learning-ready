# PYTHON LINEAR REGRESSION DEMO

### LINEAR  REGRESSION TASK

#### TRAINING DATASET

```python
# Create Linear Regression Task
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, bias=100, noise=10, random_state=123)

# Reshape Training Data
X_train, y_train = X.reshape(-1,1), y.reshape(-1,1)

# Plot Training Data
import matplotlib.pyplot as plt

plt.rcParams.update(plt.rcParamsDefault)
plt.figure(figsize=(6,4))
plt.title('Linear Regression Task')
plt.scatter(X_train, y_train, color='#000000', marker='o', s=25, alpha=0.25, label='Training Data')
plt.legend(loc='lower right')
plt.xlabel("x", fontsize=14, weight='bold')
plt.ylabel("y    ", fontsize=14, weight='bold', rotation='horizontal')
plt.xlim(-3,3)
plt.ylim(0,200)
plt.show()
```

#### FIT LINEAR MODEL

```python
from sklearn.linear_model import LinearRegression

# Fit Model
model = LinearRegression(fit_intercept=True)
model.fit(X_train, y_train)

# Print Model Coefficients
print('θ0:', model.intercept_)
print('θ1:', model.coef_)

# Use Fitted Model to Make Predictions on the whole Feature Space from -3 to +3
import numpy as np

X_space = np.linspace(-3,3,100).reshape(-1,1)
y_pred = model.predict(X_space).reshape(-1,1)

# Plot Training Data and Fitted Model
import matplotlib.pyplot as plt

plt.rcParams.update(plt.rcParamsDefault)
plt.figure(figsize=(6,4))
plt.title('Fitted Linear Model')
plt.scatter(X_train, y_train, color='#000000', marker='o', s=25, alpha=0.25, label='Training Data')
plt.plot(X_space, y_pred, color='#FC393E', linestyle='-', linewidth=2, label='Learned f(X)')
plt.axhline(model.intercept_, color='#000000', linestyle='--', linewidth=1, label='Intercept (hor.)')
plt.axvline(0, color='#000000', linestyle='--', linewidth=1, label='Intercept (vert.)')
plt.legend(loc='lower right')
plt.xlabel("x", fontsize=14, weight='bold')
plt.ylabel("y    ", fontsize=14, weight='bold', rotation='horizontal')
plt.xlim(-3,3)
plt.ylim(0,200)
plt.show()
```

### NON-LINEAR REGRESSION TASK

#### TRAINING DATASET

```python
# Create Non-linear Regression Task
import pandas as pd

X = pd.DataFrame(np.linspace(-3, 3, 100) + np.random.randn(100)/5)
X.columns=['x']
X['x_poly'] = X['x']**2
y = X['x']**2 + np.random.randn(100)*0.5 + 5

# Rename Training Data
X_train, y_train = X, y

# Plot Training Data
import matplotlib.pyplot as plt

plt.rcParams.update(plt.rcParamsDefault)
plt.figure(figsize=(6,4))
plt.title('Non-Linear Regression Task')
plt.scatter(X_train['x'], y_train, color='#000000', marker='o', s=25, alpha=0.25, label='Training Data')
plt.legend(loc='upper right')
plt.xlabel("x", fontsize=14, weight='bold')
plt.ylabel("y    ", fontsize=14, weight='bold', rotation='horizontal')
plt.xlim(-4,4)
plt.ylim(0,20)
plt.show()
```

#### FIT NON-LINEAR MODEL

```python
from sklearn.linear_model import LinearRegression

# Fit Model
model = LinearRegression(fit_intercept=True)
model.fit(X_train, y_train)

# Print Model Coefficients
print('θ0:', model.intercept_)
print('θ1, θ2:', model.coef_)

# Use Fitted Model to Make Predictions on the whole Feature Space from -3 to +3
import numpy as np

X_space = pd.DataFrame(np.linspace(-3, 3, 100) + np.random.randn(100)/5)
X_space.columns=['x']
X_space['x_poly'] = X_space['x']**2
y_pred = model.predict(X_space)

# Plot Training Data and Fitted Model
import matplotlib.pyplot as plt

plt.rcParams.update(plt.rcParamsDefault)
plt.figure(figsize=(6,4))
plt.title('Fitted Non-linear Model')
plt.scatter(X_train['x'], y_train, color='#000000', marker='o', s=25, alpha=0.25, label='Training Data')
plt.plot(X_space['x'], y_pred, color='#FC393E', linestyle='-', linewidth=2, label='Learned f(X)')
plt.axhline(model.intercept_, color='#000000', linestyle='--', linewidth=1, label='Intercept (hor.)')
plt.axvline(0, color='#000000', linestyle='--', linewidth=1, label='Intercept (vert.)')
plt.legend(loc='upper right')
plt.xlabel("x", fontsize=14, weight='bold')
plt.ylabel("y    ", fontsize=14, weight='bold', rotation='horizontal')
plt.xlim(-4,4)
plt.ylim(0,20)
plt.show()
```
